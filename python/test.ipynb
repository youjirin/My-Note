```python
import requests 
import json

res = requests.get('http://feed.mix.sina.com.cn/api/roll/get?pageid=48&lid=25&num=13&versionNumber=1.2.8&page=1&encode=utf-8&callback=feedCardJsonpCallback&_=1516515107514').text
comments = res.strip('try{').strip('feedCardJsonpCallback(').rstrip('{};').rstrip('catch(e)').rstrip('};').rstrip(')')
jd = json.loads(comments)
news = jd['result']['data']
title = [ new.get('title') for new in news]
title
    
```
